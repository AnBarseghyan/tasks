### harvest mean for different value of the feature

```python
#analysis of harvest_mean by different value of the feature
def plot_features_dist(df, column_name):
    plot_table = {}
    perc = -1
    for i in range(0, 100, 10):
        limit_down = np.percentile(df[column_name], i)
        limit_up = np.percentile(df[column_name], i+10)
        if limit_down > perc:
            plot_table[i] = int(df[(df[column_name] >= limit_down) & 
                                    (df[column_name] <= limit_up)]['harvest_mean'].mean())
            perc = limit_down
    courses = list(plot_table.keys())
    values = list(plot_table.values())
  
    fig, ax = plt.subplots()
    
    bar_height = list(plot_table.values())
    bar_x = list(range(1, len(bar_height)+1))
    bar_tick_label = list(plot_table.keys())
    bar_label = list(plot_table.values())

    bar_plot = plt.bar(bar_x,bar_height,tick_label=bar_tick_label)

    def autolabel(rects):
        for idx,rect in enumerate(bar_plot):
            height = rect.get_height()
            ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                    bar_label[idx],
                    ha='center', va='bottom', rotation=0)

    autolabel(bar_plot)

    plt.ylim(50,300)

    plt.title(f'harvest_mean for different value of {column_name}')
    plt.ylabel('harvest_mean')
    plt.xlabel('percentile')

    plt.savefig(f"{column_name}.png")
    plt.show()
```

```python
def scaling(data):
    scale_norm = StandardScaler()
    return scale_norm.fit(data)
```

```python
def pca_method(data, for_index, components=2):
    pca = PCA(n_components=components)
    principalComponents = pca.fit_transform(data)
    principalDf = pd.DataFrame(data = principalComponents, index=for_index)
    return pca, principalDf
```

```python
def plot_pca(pca):
    plt.figure(figsize=(4, 4))
    var = np.round(pca.explained_variance_ratio_*100, decimals = 1)
    lbls = [str(x) for x in range(1,len(var)+1)]
    plt.bar(x=range(1,len(var)+1), height = var, tick_label = lbls)
    print(var.sum())
    plt.show()
```

```python
scale_norm_plant = scaling(df_planter)
X_plant = scale_norm_plant.fit_transform(df_planter)
pca_plant, principalDf_plant = pca_method(X_plant, df_planter.index, 3)
```
